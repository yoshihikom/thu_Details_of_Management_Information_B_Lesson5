方針が決まったら、試行錯誤しながら、コードを書く。<br>
最終的なコードを以下に示す。

```python
import urllib.request
import re
import csv

url = 'http://koko-pachi.com/htm/korona.tokyo.6.htm'
with urllib.request.urlopen(url) as f:
    #print(f.read().decode('cp932'))
    html = f.read().decode('cp932')
    #print(html)
#file = open('test.txt', 'a')
#file.write(html)
# スクレイピングした情報を保存するlistを初期化
corona_pachinko_status_list = [["番号","住所","ホール名","状態"]]
# 読み込んだhtmlから、店舗情報の部分を抜き出して、情報を取り出すことを繰り返す
for pachinko_info in re.findall(r'right(.*?)</td></tr><tr><td nowrap align=', html, re.DOTALL):
    print(pachinko_info,"\n")
    # 抜き出した店舗情報から、まず番号を取り出す
    if re.search(r'>(.*?)</td><td nowrap><table border=1>',pachinko_info,re.DOTALL):
        number = int(re.search(r'>(.*?)</td><td nowrap><table border=1>',pachinko_info,re.DOTALL).group(1))
        print(number)
        # 次に、住所を取り出す。住所にビル名がある場合はそれをプラスする
        address = re.search(r'title="Google Map" target=_blank>(.*?)</a>',pachinko_info,re.DOTALL).group(1)
        if re.search(r'title="Google Map" target=_blank>.*?</a></td><td>(.*?)</td>',pachinko_info,re.DOTALL):
            address2 = re.search(r'title="Google Map" target=_blank>.*?</a></td><td>(.*?)</td>',pachinko_info,re.DOTALL).group(1)
            address = address + address2
        print(address)
        # ホール名を取り出す。
        if re.search(r'title="HP" target=_blank>(.*?)</a></td>',pachinko_info,re.DOTALL):
            name = re.search(r'title="HP" target=_blank>(.*?)</a></td>',pachinko_info,re.DOTALL).group(1)
        elif re.search(r'</td></tr></table></td></tr><tr><td nowrap><table border=1><tr><td>(.*?)</td>',pachinko_info,re.DOTALL):
            name = re.search(r'</td></tr></table></td></tr><tr><td nowrap><table border=1><tr><td>(.*?)</td>',pachinko_info,re.DOTALL).group(1)
        print(name)
        # 状態を取り出す。
        status = re.search(r'<tr bgcolor=.*?><td>(.*?)</td></tr></table>',pachinko_info,re.DOTALL).group(1)
        print(status,"\n") 
        # 取り出した情報を、listに保存する
        corona_pachinko_status_list.append([number,address,name,status])
print(corona_pachinko_status_list)
# listをCSVファイルに書き出す
f = open('corona_pachinko_status.csv', 'w', newline='')
writer = csv.writer(f)
writer.writerows(corona_pachinko_status_list)
f.close()
                         
```
